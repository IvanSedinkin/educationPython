```python
import pandas as pd
import numpy as np
```

```python
#read_data
df_partisipans = pd.read_csv('../archive/dataset_olympics.csv')
df_partisipans.head(10)
```

```python
# missing values count
missing_values_count = df_partisipans.isnull().sum()
missing_values_count
```

```python
# count all missing values %
data_total = np.product(df_partisipans.shape)
missing_total = missing_values_count.sum()
percent_missing = (missing_total/data_total)*100
f'Total percent of missing data = {percent_missing}'
```

```python
# drop strings with null values 
df_partisipans.dropna()
```

```python
# to much...
df_not_null = df_partisipans.dropna()
df_not_null.shape
```

```python
# drop columns with null values
df_not_null_data = df_partisipans.dropna(axis=1)
df_not_null_data
```

```python
# now its better
df_not_null_data.isnull().sum()
```

```python
df_not_null_data.head() 
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % df_partisipans.shape[1])
print("Columns with na's dropped: %d" % df_not_null_data.shape[1])
```

```python
# can also replase null with something
df_replased_data = df_partisipans
df_replased_data['Medal'] = df_partisipans['Medal'].fillna('none')
df_replased_data
```

```python
# and also can try to autofill data jast for expiriment
# we can take next row for empty value as new value, but this case not so good for this
# reason - all people are diffirent
df_replased_data.fillna(method='bfill', axis=0).fillna(0)
```
